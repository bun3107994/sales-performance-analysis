```python
import pandas as pd

df = pd.read_csv('../data/raw/train.csv')
print(f"Shape: {df.shape}")
print("Data loaded successfully!")
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)

print("Date columns fixed!")
print(df[['Order Date', 'Ship Date']].dtypes)
```

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Month Name'] = df['Order Date'].dt.strftime('%B')
df['Month-Year'] = df['Order Date'].dt.to_period('M')

print("New columns created!")
df[['Order Date', 'Year', 'Month', 'Month Name', 'Month-Year']].head()
```

```python
df = df.drop(columns=['Row ID', 'Postal Code', 'Country'])

print(f"Remaining columns: {df.shape[1]}")
print(df.columns.tolist())
```

```python
df.to_csv('../data/cleaned/clean_sales.csv', index=False)
print("Clean file saved successfully!")
```
